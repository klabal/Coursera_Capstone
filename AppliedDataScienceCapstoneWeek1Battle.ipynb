# Applied Data Science Capstone project. The Battle of the Neighborhoods (Week 1) 

## 1. Introduction. Business proposal

The capstone focuses on comparing two major cities - London, UK and New York City, NY, USA in regards to their attractiveness for prospective tourists. 
For someone not knowing much about either city, exploring as much as possible is essential. However, some things can be considered as guidelines for attractiveness towards tourists - restaurants, cafes, entertainment places like museums, theaters, pubs/bars, art galleries, hotels, stadiums, arenas. But how does one (given a positive financial and pandemic situation) decide between the two cities for a short visit? The essence of the project is to give a clear answer to the aforementioned question with the data about the key parameters (entertainment venues, restaurants, hotels, cafes). The aim of the project is to 1) use k-means clustering approach to building the model of clusters, 2) to utilize Foursquare API in getting venue data for the analysis.

## 2. Data and resources

London's borough and neighborhood data is scraped from Wikipedia https://en.wikipedia.org/wiki/List_of_areas_of_London

The features are limited to:
borough - name of the borough
postal code (postcode district) - postal code for London
post town - the town.
Towns that have the word "London" in them, will be used for the dataframe.

The data for New York City will be scraped from the json file provided during the lab assignment: https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/labs/newyork_data.json

The features are limited to:
borough - name of the borough
neighbrhood - neighborhood name
latitude - latitude coordinates
longitude - longitude coordinates

Due to the lack of geographical data of London, latitude and longitude can be obtained with the help of the geocode geopy module in Python with the help of the open-source OpenStreetMap Nominatim service.

Entertainment venue, restaurant and cafe, hotel data can be obtained via Foursquare API. The data obtained from
Foursquare will be used to utilize the maps of restaurants and cafes, entertainment venues, and hotels. For each neighborhood, the radius of analysis shall be 150 meters. For each venue, restaurant, cafe or hotel, the information will consist of:
1. Neighborhood;
2. Neighborhood latitude;
3. Neighborhood longitude;
4. Name of attraction point (might be a venue, restaurant, cafe, or hotel)
5. Attraction point latitude;
6. Attraction point longitude.

With the data collected, preprocessed, cleaned, and deemed sufficient enough for analysis, the model will then be built. 

Libraries needed for the project: numpy, pandas, folium, iPython display, scikit learn, json, xml, geocoder, beautiful soup, requests, matplotlib.
